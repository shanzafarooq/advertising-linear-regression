```python
Q1=df["Newspaper"].quantile(0.25)
print("lower quantile:",Q1)
Q3=df["Newspaper"].quantile(0.75)
print("upper quantile:",Q3)
```

```python
IQR=Q3-Q1
```

```python
IQR
```

```python

Lower_Limit = Q1 - 1.5 * IQR
Upper_Limit = Q3 + 1.5 * IQR

print("Lower Limit =", Lower_Limit)
print("Upper Limit =", Upper_Limit)
```

```python
outliers = df[(df["Newspaper"] < Lower_Limit) | (df["Newspaper"] > Upper_Limit)]

print("Lower Limit:", Lower_Limit)
print("Upper Limit:", Upper_Limit)
```

```python
X = df[["TV", "Radio", "Newspaper"]]
y = df["Sales"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=4)
```

```python
y
```

```python
x_train.shape
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_test
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2score=r2_score(y_test,y_predict)
```

```python
r2score
```

```python
import math
print(math.sqrt(mean_squared_error(y_test,y_predict)))
```

```python
print(math.sqrt(mean_absolute_error(y_test,y_predict)))
```
